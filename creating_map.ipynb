```python
import pandas as pd
import numpy as np
from plotly import graph_objects as go
from plotly.subplots import make_subplots
import folium
from folium import IFrame
import branca
import os
import geopandas
```

```python
df = pd.read_excel('locaties.xlsx')

df['colour']= (df['html_file'].str.len() < 8)
df['colour'] = df['colour'].replace(True,'red')
df['colour'] = df['colour'].replace(False,'green')
```

```python
m = folium.Map(location=["51.946503", "5.817492"],zoom_start=12)

for a,b in df.iterrows():
    html="""
    <iframe src=\"""" + b['html_file'] + """\" width="850" height="400"  frameborder="0">    
    """
    
    popup = folium.Popup(folium.Html(html, script=True))
    folium.Marker([b['longitude'],b['latitude']],popup=popup,icon=folium.Icon(icon='flash', prefix='fa',color=b['colour'])).add_to(m)
```
